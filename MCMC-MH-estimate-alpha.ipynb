```python
#1000 sample 5000 iteration
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import invgamma
from scipy.stats import norm
from scipy.stats import levy_stable

alpha = 1.5 
beta = 0
gamma = 2.5
sigma = 1.0
samples = 1000
y = levy_stable.rvs(alpha = alpha, beta = beta, loc = gamma, scale = sigma, size = samples)

T = 5000
alpha = [0.0] * (T + 1)
sigma = [0.0] * (T + 1)
gamma = [0.0] * (T + 1)
#initial
alpha[0] = 1.2
sigma[0] = 1.0
gamma[0] = 2.5
print(len(alpha),len(sigma),len(gamma))
#prior
p_alpha = uniform(0, 2)
p_sigma = invgamma(a = 1)
p_gamma = norm(loc = np.mean(y), scale = 1/0.2)
print(p_alpha,p_sigma,p_gamma)

t = 0
while t < T:
    t = t + 1
    print("this is %d iteration"%t)
    alpha_new = p_alpha.rvs(1)
    sigma_new = p_sigma.rvs(1)
    gamma_new = p_gamma.rvs(1)
    # step 1
    pn_po1 = 1
    for i in range(len(y)):
        pn_po1 = pn_po1 * levy_stable.pdf(x = y[i], alpha = alpha_new, beta = 0, loc = gamma[t-1], scale = sigma[t-1])/levy_stable.pdf(x = y[i], alpha = alpha[t-1], beta = 0, loc = gamma[t-1], scale = sigma[t-1])
    print(".....",pn_po1)
    A_alpha = min(1, pn_po1)
#     print(A_alpha)

    u = random.uniform(0, 1)
    if u <= A_alpha:
        alpha[t] = alpha_new
    else:
        alpha[t] = alpha[t - 1]


    # step 2
    pn_po2 = 1
    for i in range(len(y)):
        pn_po2 = pn_po2 * levy_stable.pdf(x = y[i], alpha = alpha[t-1], beta = 0, loc = gamma_new, scale = sigma[t-1])/levy_stable.pdf(x = y[i], alpha = alpha[t-1], beta = 0, loc = gamma[t-1], scale = sigma[t-1])
    print("++++",pn_po2)
    A_gamma = min(1, ((pn_po2 * p_gamma.pdf(gamma_new))/(p_gamma.pdf(gamma[t-1]))))

    u = random.uniform(0, 1)
    if u <= A_gamma:
        gamma[t] = gamma_new
    else:
        gamma[t] = gamma[t - 1]
        
    # step 3
    pn_po3 = 1
    for i in range(len(y)):
        pn_po3 = pn_po3 * levy_stable.pdf(x = y[i], alpha = alpha[t-1], beta = 0, loc = gamma[t-1], scale = sigma_new)/levy_stable.pdf(x = y[i], alpha = alpha[t-1], beta = 0, loc = gamma[t-1], scale = sigma[t-1])
    print("====",pn_po3)
    A_sigma = min(1, ((pn_po3 * p_sigma.pdf(sigma_new))/(p_sigma.pdf(sigma[t-1]))))

    u = random.uniform(0, 1)
    if u <= A_sigma:
        sigma[t] = sigma_new
    else:
        sigma[t] = sigma[t - 1]
        



```

```python
plt.figure(figsize=(6, 3))

plt.plot(range(T + 1), alpha, 'steelblue',linestyle='-', label="0")
plt.ylabel(" value")
plt.xlabel("Iterations")

```
